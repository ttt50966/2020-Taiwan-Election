```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1_后里 = pd.read_csv("后里區_政黨票得票率.txt")
df2_后里 = pd.read_csv("后里區_區域立委得票率.txt")
df_后里= df1_后里[df1_后里.columns[1:]]
for 候選人 in df2_后里.columns[1:-1]:
    df_后里[候選人]=df2_后里[df2_后里.columns[1:]][候選人]

df_中三選區 = df_后里
區域 = ["潭子區", "神岡區", "大雅區"]
for 區 in 區域:
    df1 = pd.read_csv("%s_政黨票得票率.txt" %(區))
    df2 = pd.read_csv("%s_區域立委得票率.txt" %(區))
    df= df1[df1.columns[1:]]
    for 候選人 in df2.columns[1:-1]:
        df[候選人]=df2[df2.columns[1:]][候選人]
    df_中三選區 = df_中三選區.append(df, ignore_index=True)
df_中三選區.corr().style.background_gradient(cmap='coolwarm')
```

```python
import imgkit
html = df_中三選區.corr().style.background_gradient(cmap='coolwarm').set_precision(2).render()
with open("my.html","w") as fp:
   fp.write(html)
```
