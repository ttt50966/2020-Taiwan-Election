# 台中第二選區

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 沙鹿

```python
df1 = pd.read_csv("沙鹿區_政黨票得票率.txt")
df2 = pd.read_csv("沙鹿區_區域立委得票率.txt")
df_沙鹿 = df1[df1.columns[1:]]
df_沙鹿["顏寬恆"]=df2[df2.columns[1:]]["顏寬恆"]
df_沙鹿["陳柏惟"]=df2[df2.columns[1:]]["陳柏惟"]
df_沙鹿.corr().style.background_gradient(cmap='coolwarm')
```

## 烏日

```python
df1 = pd.read_csv("烏日區_政黨票得票率.txt")
df2 = pd.read_csv("烏日區_區域立委得票率.txt")
df_烏日 = df1[df1.columns[1:]]
df_烏日["顏寬恆"]=df2[df2.columns[1:]]["顏寬恆"]
df_烏日["陳柏惟"]=df2[df2.columns[1:]]["陳柏惟"]
df_烏日.corr().style.background_gradient(cmap='coolwarm')
```

## 綜合全部

```python
#沙鹿 烏日 大肚 龍井 霧峰
df_中二選區 = df_沙鹿
df_中二選區 = df_中二選區.append([df_烏日,df_大肚,df_龍井,df_霧峰],ignore_index=True)
df_中二選區.corr().style.background_gradient(cmap='coolwarm')
```
