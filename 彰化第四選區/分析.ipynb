## 永靖鄉

```python
df1 = pd.read_csv("永靖鄉_政黨票得票率.txt")
df2 = pd.read_csv("永靖鄉_區域立委得票率.txt")
df= df1[df1.columns[1:]]
for 候選人 in df2.columns[1:]:
    df[候選人]=df2[df2.columns[1:]][候選人]
df.corr().style.background_gradient(cmap='coolwarm')
```

## 綜合區域

```python
df1_永靖 = pd.read_csv("永靖鄉_政黨票得票率.txt")
df2_永靖 = pd.read_csv("永靖鄉_區域立委得票率.txt")
df_永靖= df1_永靖[df1_永靖.columns[1:]]
for 候選人 in df2_永靖.columns[1:]:
    df_永靖[候選人]=df2_永靖[df2_永靖.columns[1:]][候選人]

df_彰四選區 = df_永靖
區域 = ["二水鄉","員林市", "社頭鄉", "田中鎮", "田尾鄉", "大村鄉"]
for 區 in 區域:
    df1 = pd.read_csv("%s_政黨票得票率.txt" %(區))
    df2 = pd.read_csv("%s_區域立委得票率.txt" %(區))
    df= df1[df1.columns[1:]]
    for 候選人 in df2.columns[1:]:
        df[候選人]=df2[df2.columns[1:]][候選人]
    df_彰四選區 = df_彰四選區.append(df, ignore_index=True)
df_彰四選區.corr().style.background_gradient(cmap='coolwarm')
```

```python
import imgkit
html = df_彰四選區.corr().style.background_gradient(cmap='coolwarm').set_precision(2).render()
imgkit.from_string(html, '彰四選區.png', options= {'encoding': "UTF-8"})
```
